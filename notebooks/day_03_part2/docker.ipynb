### What is a container?

Its like the house built from that blueprint. It’s the running instance where the app actually lives and works

### Why do we use containers?

- Portability: Runs the same everywhere.
- Efficiency: Faster and uses fewer resources than VMs.
- Scalability: Easy to spin up/down multiple copies.
- Consistency: Developers, testers, and production all use the same setup.

### What is a docker image?

like an installation package or a blueprint. It’s contains all the instructions (what software, dependencies, configs, etc.).

### Explain the RUN and ENV lines you added to the file

- apt-get update → updates package lists
- apt-get install -y cowsay → installs cowsay without prompting
- apt-get clean && rm -rf /var/lib/apt/lists/* → cleans up temporary files to keep the image small
- Docker looks for cowsay, in /usr/games/cowsay

```python
# build the docker image
!docker build -t my-cowsay cowsay-docker
```

```python
# make sure that the image has been built
!docker image ls
```

## Do you think bioinformaticians have to create a docker image every time they want to run a tool?
